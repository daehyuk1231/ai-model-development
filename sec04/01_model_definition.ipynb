##### 임포트

```python
import torch
import torch.nn as nn
```

##### Device 설정

```python
# GPU(CUDA) 사용 가능 시 'cuda', 아니면 'cpu' 사용
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"사용 장치: {device}")
```

##### FC 층 (완전 연결층, Fully Connected Layer, Dense Layer) 정의

```python
class FCLayerModel(nn.Module):
    def __init__(self):
        # nn.Module의 초기화 메서드 호출 (필수)
        super().__init__()

        # 완전 연결 레이어(Fully Connected Layer) 정의
        # 입력 특성 10개
        # 출력 특성  2개
        # 10개를 2개로 분류
        self.fc = nn.Linear(10, 2)

    # 순전파 메소드
    # - 입력 x를 받아서 완전 연결 레이어를 통과시킴
    def forward(self, x):
        return self.fc(x)

# 모델 인스턴스 생성
model = FCLayerModel().to(device)
print(model)
```

##### DNN(DNN, Deep Neural Network, 은닉층을 가진 신경망) 정의

```python
class DnnModel(nn.Module):
    def __init__(self):
        super().__init__()

        # 입력특성(10) → 출력특성(64)
        self.fc1 = nn.Linear(10, 64)

        # 입력특성(64) → 출력특성(32)
        self.fc2 = nn.Linear(64, 32)

        # 입력특성(32) → 출력특성(2)
        self.fc3 = nn.Linear(32, 2)

    # 순전파 메소드
    def forward(self, x):
        # 첫 번째 은닉층 - 저수준 특징 추출
        x = self.fc1(x)

        # 두 번째 은닉층 - 고수준 특징 추출
        x = self.fc2(x)

        # 출력층 - 이진 분류
        x = self.fc3(x)

        return x

# 모델 인스턴스 생성
model = DnnModel().to(device)
print(model)
```

##### Sequential로 DNN 모델 정의

```python
class DnnModel2(nn.Module):
    def __init__(self):
        super().__init__()

        # nn.Sequential: 레이어를 순차적으로 연결
        # forward 메서드에서 순서대로 자동 실행됨
        self.model = nn.Sequential(
            # 입력 특성(10) → 출력 특성(64)
            nn.Linear(10, 64),

            # 입력 특성(64) → 출력 특성(32)
            nn.Linear(64, 32),

            # 입력 특성(32) → 출력 특성(2)
            nn.Linear(32, 2)
        )

    # 순전파 메소드
    def forward(self, x):
        # self.model에 정의된 레이어 순서대로 연산
        return self.model(x)

# 모델 인스턴스 생성
model = DnnModel2().to(device)
print(model)
```
