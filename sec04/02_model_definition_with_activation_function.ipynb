##### 임포트

```python
import torch
import torch.nn as nn
```

##### Device 설정

```python
# GPU(CUDA) 사용 가능 시 'cuda', 아니면 'cpu' 사용
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"사용 장치: {device}")
```

##### 출력 뉴런에 활성화 함수 적용

```python
class DnnWithActivation(nn.Module):
    def __init__(self):
        super().__init__()

        # nn.Sequential: 레이어를 순차적으로 연결
        self.model = nn.Sequential(
            nn.Linear(10, 64),
            nn.ReLU(),          # ReLU 활성화 함수: 음수는 0, 양수는 그대로(뉴런마다 실행)
            nn.Linear(64, 32),
            nn.ReLU(),          # ReLU 활성화 함수: 음수는 0, 양수는 그대로(뉴런마다 실행)
            nn.Linear(32, 2)    # 출력
        )

    # 순전파 메소드
    def forward(self, x):
        # self.model에 정의된 레이어 순서대로 연산
        return self.model(x)

# 모델 인스턴스 생성
model = DnnWithActivation().to(device)
print(model)
```
