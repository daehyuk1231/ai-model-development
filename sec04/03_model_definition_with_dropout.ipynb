##### 임포트

```python
import torch
import torch.nn as nn
```

##### Device 설정

```python
# GPU(CUDA) 사용 가능 시 'cuda', 아니면 'cpu' 사용
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"사용 장치: {device}")
```

##### 드롭아웃 적용 모델

```python
class DnnWithDropout(nn.Module):
    def __init__(self):
        super().__init__()

        self.model = nn.Sequential(
            nn.Linear(10, 64),
            nn.ReLU(),
            nn.Dropout(0.3),       # 64뉴런에서 30%를 무작위로 비활성화(0으로 출력)

            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Dropout(0.3),       # 32뉴런에서 30%를 무작위로 비활성화(0으로 출력)

            nn.Linear(32, 2)       # 최종 출력 (드롭 아웃은 출력층에서는 일반적으로 사용하지 않음)
        )

    # 순전파 메서드 정의
    def forward(self, x):
        return self.model(x)

# 모델 인스턴스 생성
model = DnnWithDropout().to(device)
print(model)
```
