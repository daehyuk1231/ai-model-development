##### 임포트

```python
import numpy as np
import torch
```

##### Numpy ndarray의 수동 미분

```python
# ndarray 객체 생성
x_np = np.array(3.0)

# 연산 수행
y_np = 2 * x_np

# 순간 변화율(기울기, 미분값, dy/dx)을 수동으로 계산
dx = 1e-5
y_plus_dx = 2 * (x_np + dx)
gradient = (y_plus_dx - y_np) / dx
print(f"기울기(dy/dx) = {gradient}") # 복잡한 수식에서 오차가 발생할 수 있음
```

##### Pytorch Tensor의 자동 미분

```python
# Tensor 객체 생성:
# - requires_grad=True: 이 옵션이 활성화되면 연산이 수행될 때마다 내부적으로 연산 과정을 기록
x_pt = torch.tensor(3.0, requires_grad=True)

# 연산 수행 → 계산 과정 기록(계산 그래프 생성)
y_pt = 2 * x_pt

# 순간 변화율 (기울기, 미분값, dy/dx)을 역전파하면서 자동 계산
y_pt.backward()

print(f"기울기(dy/dx) = {x_pt.grad}")
```
