##### 임포트

```python
import torch
import numpy as np
```

##### 파이썬 리스트 -> 텐서

```python
# 리스트
label = [0, 1, 2, 3]

# 리스트 -> 텐서
label_tensor = torch.tensor(label, dtype=torch.float32)

print(f"타입: {type(label_tensor)}, 형태: {label_tensor.shape}, 데이터 타입: {label_tensor.dtype}")
label_tensor
```

##### 넘파이 배열 -> 텐서

```python
# 넘파이 배열
label = np.array([0, 1, 2, 3])

# 넘파이 배열 -> 텐서
label_tensor = torch.tensor(label, dtype=torch.float32)

print(f"타입: {type(label_tensor)}, 형태: {label_tensor.shape}, 데이터 타입: {label_tensor.dtype}")
label_tensor
```
