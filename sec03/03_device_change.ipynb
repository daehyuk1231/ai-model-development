##### 임포트

```python
import torch
```

##### 텐서를 GPU 메모리로 복사 이동 방법1

```python
# 텐서를 CPU 메모리에 생성
# torch.rand(1000, 1000):
# - 0~1 사이의 균등분포에서 난수 생성, 2차원 텐서(1000x1000) 생성
tensor1 = torch.rand(1000, 1000)
print(f"tensor1.device: {tensor1.device}")

# GPU가 사용 가능한 경우, tensor1을 GPU 메모리에 복사
if torch.cuda.is_available():
    # 원본 tensor1은 CPU에 그대로 유지하고
    # tensor1을 GPU 메모리에 복사해서 새로 텐서 생성
    tensor2 = tensor1.to('cuda') # tensor1.cuda()
    print(f"tensor2.device: {tensor2.device}")

```

##### 텐서를 GPU 메모리로 복사 이동 방법2

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# 텐서를 CPU 메모리에 생성
tensor1 = torch.rand(1000, 1000)

# GPU가 사용 가능한 경우, 텐서를 GPU 메모리로 이동
tensor2 = tensor1.to(device)

print(f"tensor1.device: {tensor1.device}")
print(f"tensor2.device: {tensor2.device}")
```

##### 텐서를 GPU 메모리로 복사 이동 방법3

```python
# 디바이스 결정
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# 텐서를 처음부터 GPU 메모리에 생성
tensor = torch.rand(1000, 1000, device=device)

print(f"tensor.device: {tensor.device}")
```
