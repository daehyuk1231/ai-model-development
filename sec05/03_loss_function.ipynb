##### 임포트

```python
import torch
import torch.nn as nn
```

##### Device 설정

```python
# GPU(CUDA) 사용 가능 시 'cuda', 아니면 'cpu' 사용
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"사용 장치: {device}")
```

##### 손실 함수

```python
# 3종 품종 분류: CrossEntropyLoss 사용
loss_fn = nn.CrossEntropyLoss()
```

```python

# 샘플 3개에 대해서 3종 품종을 분류하는 모델 출력
model_outputs = torch.tensor([
    [2.0, 1.0, 0.5],   # 샘플1: 품종A에 높은 점수
    [0.6, 0.4, 2.1],   # 샘플2: 품종C에 높은 점수
    [0.3, 2.5, 0.8]    # 샘플3: 품종B에 높은 점수
], device=device)

# 정답 레이블
targets = torch.tensor([0, 2, 1], device=device) 

# 손실 계산: 모델의 출력과 정답이 얼마나 다른지를 수치로 나타낸 값
loss = loss_fn(model_outputs, targets)
print(f"손실: {loss.item():.4f}") 

# 예측 확률로 변환(소프트맥스)
probabilities = model_outputs.softmax(dim=1)
print(f"예측 확률:{probabilities}")

# 예측된 클래스 얻기: 가장 높은 확률을 가진 클래스 선택
predicted_class = probabilities.argmax(dim=1)
print(f"예측 클래스: {predicted_class}")
print(f"정답 클래스: {targets}")

# 정확도 계산
accuracy = (predicted_class == targets).float().mean().item() * 100
print(f"정확도: {accuracy:.2f}%")
```
