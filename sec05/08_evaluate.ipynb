##### 평가하기

```python
# 평가 모드로 전환
model.eval()

# 역전파/기울기 계산없이 테스트셋으로 예측 수행
with torch.inference_mode():  # torch.no_grad():
    # 모델 출력 얻기
    test_outputs  = model(test_x_tensor)
    # 테스트 손실 계산
    test_loss     = loss_fn(test_outputs, test_y_tensor)    
    # 테스트 정확도 계산
    test_correct  = (test_outputs.argmax(dim=1) == test_y_tensor).sum().item()
    test_acc = test_correct / len(test_x_tensor) * 100

print(f"테스트 손실:   {test_loss.item():.4f}  "
      f"테스트 정확도: {test_acc:.1f}% ")
```
